```python
from transformers import AutoTokenizer
from transformers import AutoModelForMultipleChoice
import torch
```

```python
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModelForMultipleChoice.from_pretrained("bert-base-uncased")
```

```python
prompt = ["Amphibians are four-limbed and ectothermic vertebrates of the class Amphibia. A frog is an amphibian. A frog is any member of a diverse and largely carnivorous group of short-bodied, tailless amphibians composing the order Anura. The oldest fossil 'proto-frog' Triadobatrachus is known from the Early Triassic of Madagascar, but molecular clock dating suggests their split from other amphibians may extend further back to the Permian, 265 million years ago.."]
candidates = ["An animal is a frog.","A frog is an amphibian.","A frog is an animal."]
```

```python
from itertools import cycle

data = [[x,y] for x,y in zip(cycle(prompt),candidates)]
inputs = tokenizer(data, return_tensors="pt", padding=True)
labels = torch.tensor(0).unsqueeze(0)
data
```

```python
outputs = model(**{k: v.unsqueeze(0) for k, v in inputs.items()}, labels=labels)
```

```python
logits = outputs.logits
```
